# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Import the Dataset

```python
df = pd.read_csv('Position_Salaries.csv')
X = df.iloc[:, 1:-1].values  # Feature matrix (2D array)
y = df.iloc[:, -1].values  # Target vector (1D array)
```

```python
# df
```

```python
# X
```

# Train Linear Regression Model On the Whole Dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
# Train Linear Regression Model On the Whole Dataset to maximize accuracy
lin_reg.fit(X, y)
```

# Train Polynomial Linear Regression Model On the Whole Dataset

```python
from sklearn.preprocessing import PolynomialFeatures
# Train Polynomial Linear Regression Model On the Whole Dataset
poly_reg = PolynomialFeatures(degree = 4)  # Degree of the polynomial (2 and 3 are common choices but didn't work well here)
# Transform X to polynomial features
X_poly = poly_reg.fit_transform(X)  # Transform X to polynomial features
# Create a new Linear Regression model
lin_reg_2 = LinearRegression()
# Fit the linear regression model to the polynomial features
lin_reg_2.fit(X_poly, y)    
```

# Visualizing the Linear Regression Results

```python
# Visualizing the Linear Regression Results
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

# Visualizing the Polynomial Linear Regression Results

```python
# Visualizing the Polynomial Linear Regression Results
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(X_poly), color = 'blue')
plt.title('Truth or Bluff (Polynomial Linear Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

# Visualizing the Polynomial Linear Regression Results (High resolution, Smooth curve)

```python
# Visualizing the Polynomial Linear Regression Results (High resolution, Smooth curve)
X_grid = np.arange(min(X), max(X), 0.1)  # Create a high-resolution grid
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Linear Regression - High Resolution)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

# Predicting a New Result With Linear Regression

```python
# Predicting a New Result With Linear Regression
lin_reg.predict([[6.5]])
```

# Predicting a New Result With Polynomial Linear Regression

```python
# Predicting a New Result With Polynomial Linear Regression
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
